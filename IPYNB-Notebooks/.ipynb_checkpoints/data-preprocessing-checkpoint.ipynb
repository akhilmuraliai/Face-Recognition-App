## DATA PREPARATION

#### CROPPING AND SAVING FACES USING HAARCASCADE
    1. Read the image using opencv
    2. Convert it to grayscale
    3. Apply classifier and get the coordinates of faces
    4. Save the face by appying the ROI
    
  

```python
def crop_face(path, gender, i):
    
    img = cv2.imread(path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    harr = cv2.CascadeClassifier('./harrcascade-classifier/haarcascade_frontalface_default.xml')
    faces = harr.detectMultiScale(gray, 1.5, 5)
    
    for x, y, w, h in faces:
        
        roi = img[y:y+h, x:x+w]
        
        if gender == 'male':
            cv2.imwrite('./image-data/crop/male-crop/{}_{}.png'.format(gender, i), roi)
        else:
            cv2.imwrite('./image-data/crop/female-crop/{}_{}.png'.format(gender, i), roi)          
```

```python
error_in_female = []

for i, path in enumerate(female_images_path):
    try:
        crop_face(path, 'female', i)
#         print(f'{i}/{len(female_images_path)} processed successfully')
    except:
        error_in_female.append(path)
        print(f'{i}/{len(female_images_path)} UNEXPECTED ERROR!')
        
        
error_in_male = []

for i, path in enumerate(male_images_path):
    try:
        crop_face(path, 'male', i)
#         print(f'{i}/{len(male_images_path)} processed successfully')
    except:
        error_in_male.append(path)
        print(f'{i}/{len(male_images_path)} UNEXPECTED ERROR!')
```

```python
print('Error in Male : ', error_in_male)
print('Error in Female : ', error_in_female)
```

#### RESIZE IMAGES

```python
def resize_image(path):
    try:
        
        img = cv2.imread(path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        size = gray.shape[0]

        if size > 100:
            gray_re = cv2.resize(gray, (100, 100), cv2.INTER_AREA) #shrink
        else:
            gray_re = cv2.resize(gray, (100, 100), cv2.INTER_CUBIC) #enlarge

        return gray_re.flatten()
    
    except:
        
        return None
```

```python
print(resize_image(total_path[9]))
print(resize_image(total_path[9]).shape)
```

```python
new_df['ImageData'] = new_df['Path'].apply(resize_image)
new_df.head(10)
```
