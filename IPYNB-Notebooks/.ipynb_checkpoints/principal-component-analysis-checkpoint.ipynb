## NOTEBOOK IMPORTS AND DATA

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.decomposition import PCA

from glob import glob
from PIL import Image

import warnings
warnings.filterwarnings('ignore')
```

```python
data = np.load('./models/Xy-preprocessed.npz')
data.files
```

```python
X = data['arr_0']
y = data['arr_1']

print(f'X: {X.shape}')
print(f'y: {y.shape}')
```

## EIGEN IMAGE

```python
X1 = X - X.mean(axis=0)
```

```python
pca = PCA(n_components=None, whiten=True)
x_pca = pca.fit_transform(X1)

x_pca.shape
```

```python
X1.shape
```

```python
eigen_ratio = pca.explained_variance_ratio_
eigen_rato_cum = np.cumsum(eigen_ratio)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(eigen_ratio[:200], 'r>--')
plt.xlabel('No. of components')
plt.ylabel('Explained Varaince Ratio')

plt.subplot(1, 2, 2)
plt.plot(eigen_rato_cum[:200], 'r>--')
plt.xlabel('No. of components')
plt.ylabel('Cumilative Explained Varaince Ratio')

plt.show()
```

    By using elbow method, 
        If we consider componets between 25 to 30,
        the explained variance is 75%
        
        Inorder to get minimun 80% variance, we have to consider
        50 components
        
    This means that we are only considering important features.
    Here 50 means we are taking 50 directions instead of 10000 directions

```python
pca_50_components = PCA(n_components=50, whiten=True)
x_pca_50 = pca_50_components.fit_transform(X1)

x_pca_50.shape
```

## SAVING PCA OBJECT

```python
import pickle
pickle.dump(pca_50_components, open('./models/PCA_50.pickle', 'wb'))
```

## PCA COMPONENTS INVERSE TRANSFORMS

```python
x_pca_inv = pca_50_components.inverse_transform(x_pca_50)
x_pca_inv.shape
```

#### PLOTTING A SAMPLE FROM INVERSE IMAGES

```python
eigen_image = x_pca_inv[8, :].reshape(100, 100)
plt.imshow(eigen_image, cmap='gray');
```

#### PLOTTING EIGEN IMAGES vs NORMAL IMAGES

```python
def label(y):
    if y == 0:
        return 'male'
    else:
        return 'female'
    
    
pics = np.random.randint(0, 1112, 40)

plt.figure(figsize=(15, 8))

print('-'*40 + 'Normal Images' + '-'*40)

for i, pic in enumerate(pics):
    plt.subplot(4, 10, i+1)
    img = X[pic:pic+1].reshape(100, 100)
    plt.imshow(img, cmap='gray')
    plt.title(f'{label(y[pic])}')
    plt.xticks([])
    plt.yticks([])
    plt.axis('off')
    
plt.show()
    
    
plt.figure(figsize=(15, 8))    
print('-'*40 + 'Eigen Images' + '-'*40)

for i, pic in enumerate(pics):
    plt.subplot(4, 10, i+1)
    img = x_pca_inv[pic:pic+1].reshape(100, 100)
    plt.imshow(img, cmap='gray')
    plt.title(f'{label(y[pic])}')
    plt.xticks([])
    plt.yticks([])
    plt.axis('off')

plt.show()
```

## SAVING ARRAY

```python
np.savez('./models/Data-50-Components-y-XMean.npz',
         x_pca_50, y, X.mean())
```
